```python
class DeepSeekIntegration:
    def __init__(self):
        with open('deepseek_api_key.txt', encoding='utf-8') as f:
            self.api_key = f.read()


    def call_deepseek(self, prompt: str, max_tokens: int = 1000, temperature: float = 0.8) -> str:

        headers = {
            "Authorization": f"Bearer {self.api_key}",
            "Content-Type": "application/json"
        }

        payload = {
            "model": "deepseek/deepseek-chat",
            "messages": [
                {
                    "role": "system",
                    "content": "Ты полезный ассистент, который генерирует реалистичные обращения клиентов сервиса по аренде инструментов."
                },
                {
                    "role": "user",
                    "content": prompt
                }
            ],
            "temperature": temperature,
            "max_tokens": max_tokens,
            "top_p": 0.95
        }

        response = requests.post("https://openrouter.ai/api/v1/chat/completions", headers=headers, json=payload)
        if response.status_code == 200:
            result = response.json()
            return result['choices'][0]['message']['content'].strip()

    def generate_texts_for_theme(self, theme: str, n_examples: int = 10) -> List[str]:

        prompt = f"""Ты — генератор обращений клиентов пункта проката строительного инструмента.

Клиенты чаще всего знают, какой именно инструмент им нужен, и пишут его название (например, "шуруповерт"), а иногда пишут только суть работ, которые им нужно выполнить.

Твоя задача: для темы "{theme}" сгенерировать 10 реалистичных обращений.

ПРАВИЛА:
1. Стиль: живой, разговорный, как от реального клиента (мастера, дачника, строителя, обычного человека)
2. Разнообразие ситуаций: мелкий ремонт, стройка, дача, отделка, демонтаж, земляные работы, высотные работы и т.д.
3. Длина: 1-2 предложения (10-25 слов)
4. В обращениях не должен встречаться символ точка с запятой, чтобы не ломался парсинг.
5. Нужно, чтобы название темы в обращении встречалось не чаще 30%. В остальных случаях используй синонимы из справочника ниже.

ФОРМАТ ОТВЕТА: только тексты в одинарных кавычках через точку с запятой, без нумерации, без пояснений.

СПРАВОЧНИК СНИНОНИМОВ (формат: <тема> -> <синоним темы>):
перфоратор -> перф
перфоратор -> бурильник
виброплита -> трамбовка
виброплита -> уплотнитель
виброплита -> утрамбовщик
виброплита -> нога
отбойный молоток -> отбойник
отбойный молоток -> молот
отбойный молоток -> электромолот
отбойный молоток -> стенолом
отбойный молоток -> асфальтолом
отбойный молоток -> бетонолом
генератор -> бензогенератор
генератор -> электростанция
генератор -> дизельный генератор
генератор -> гена
плиткорез -> резак по плитке
плиткорез -> станок для плитки
шуруповерт -> дрель-шуруповерт
шуруповерт -> дрель
шуруповерт -> дрель по металлу
шуруповерт -> дрель по дереву
шуруповерт -> шурик
ушм -> болгарка
ушм -> шлифмашина
ушм -> угловая шлифмашина
ушм -> шлифмашина угловая
штроборез -> штробилер
штроборез -> бороздодел
штроборез -> штрободел
штроборез -> штробилёр
штроборез -> штробарез
штроборез -> штрабарез
триммер -> косилка
триммер -> мотокоса
триммер -> бензокосилка
триммер -> бензокоса
триммер -> коса бензиновая
триммер -> коса с леской
лестница -> трансформер
лестница -> стремянка
бензобур -> ямобур
бензобур -> бурильщик
бензобур -> бурилка
бензобур -> ямобур бензиноый
скарификатор -> аэратор
жираф -> шлифовальная машина для стен и потолков
жираф -> шлифовальная машина для стен
жираф -> шлифовальная машина для потолков
мозаично-шлифовальная машина -> мозаичная машина
мозаично-шлифовальная машина -> фрезер по бетону
мозаично-шлифовальная машина -> фрезеровальная машина по бетону
мозаично-шлифовальная машина -> фрезеровальная машина для бетона
установка алмазного бурения -> установка для алмазного бурения
установка алмазного бурения -> установка для бурения
установка алмазного бурения -> алмазное бурение
торцовочная пила -> торцовка
торцовочная пила -> торцеватель
торцовочная пила -> торцовщик
торцовочная пила -> пила торцевая
торцовочная пила -> пила торцовочная
толщиномер -> толщиномер ЛКП
толщиномер -> измеритель толщины краски
толщиномер -> измеритель ЛКП
толщиномер -> измеритель краски на машине
толщиномер -> измеритель слоя краски
сварочный аппарат -> сварочник
сварочный аппарат -> сварочный полуавтомат
сварочный аппарат -> электро-дуговая сварка
сварочный аппарат -> аппарат для сварки
сварочный аппарат -> аппарат для электро дуговой сварки
сварочный аппарат -> аппарат для электро-дуговой сварки
сабельная пила -> пила сабельная
мотопомпа -> помпа
мотопомпа -> мотопомпа грязевая
мотопомпа -> бензопомпа
мотопомпа -> мото-помпа
мотопомпа -> мотопомпа для грязной воды
мотопомпа -> помпа для грязной воды
культиватор -> мотокультиватор
культиватор -> бензокультиватор
культиватор -> мотоблок
культиватор -> бензоблок
культиватор -> мотобензоблок
культиватор -> мото бензо блок

ПРИМЕРЫ ДЛЯ ТЕМЫ "шуруповерт": 'добрый вечер! подскажите, у вас можно взять в аренду шурповерт?','Здравствуйте, сколько будет стоить арендовать, шуроповёрт на одни сутки','Здравствуйте, меня зовут Александр, просили написать на Авито на прямую по поводу аренды шуруповерта и перфоратора.','Здравствуйте! Нужен шуруповерт, завтра в районе 12','Здравствуйте, мне нужно шуруповерт. На 3 дня. Какие условия?','Здравствуйте! Сколько будет стоить аренда шуруповерта?','Здравствуйте! Хотел бы уточнить стоимость аренды шуруповерта на 1 день','здравствуйте есть шуруповерт ленточный ?','Здравствуйте, подскажите пожалуйста, нам нужен в аренду шуроповерт и шестигранный ключ ( но только больше универсального, для разбора дивана). Сколько это будет стоить по цене, если арендовать его у вас на сутки, с 18:00','Здравствуйте, сколько будет стоит шуруповерт?','Здравствуйте  Аренда шуруповерта какая стоимость? Залог?','Добрый вечер! Мне нужны инструменты: шуруповёрт. У вас есть такое, если да, то сколько это будет стоить?','здравствуйте! почем аренда шуроповерта ?','Здравствуйте, нужен шуруповерт и молоток','Здравствуйте, сколько будет стоить аренда шуруповерта? И есть ли у вас возможность давать сверло?','Здравствуйте а у вас нету шуруповерта в аренду ?','Здравствуйте, сколько стоит аренда шуруповерта самого простого на сегодня','Добрый день, подскажите, пожалуйста, стоимость аренды шуруповерта?','Здравствуйте, хотел бы взять в аренду у вас 2-3 шуроповерта мощных  Нужно вкручивать саморезы через листвиницу в металл  На суббота воскресенье','Добрый день, подскажите пожалуйста, есть ли у вас шуруповёрт в прокате и сколько будет стоить его аренда на день?','здравствуйте, у вас есть в аренду дрель, шуруповерт и крестообразная ответрка?','Добрый вечер, можно ли арендовать шуруповерт ?','Здравствуйте.  Шуруповерт на сутки сколько стоит и какие условия?','Здравствуйте, есть ли у вас в наличии следующие инструменты?  шуруповерт  Нужно взять в аренду завтра в середине дня на двое суток. И сориентируйте пожалуйста по стоимости.','Здравствуйте, подскажите, можно ли сегодня подъехать арендовать шуруповерт-дрель? В районе 17:00','Здравствуйте, подскажите пожалуйста сколько будет по цене шуруповерт, на 1 день?','Добрый день, если арендовать шуруповерт, то какие биты идут в комплекте ?)','Здравствуйте! У вас есть шуруповерт в аренду?','Вечер добрый. В ближайшие дни можно арендовать шуруповерт ?','Добрый вечер  Могу взять завтра на прокат шуроповерт с насадками на крест?','Здравствуйте, сегодня можно подъехать часов в 12 взять шруповерт на сутки?','Здравствуйте, есть в аренду шуруповерт?','Здравствуйте, в какую стоимость у вас будет арендовать шуруповерт?','здравствуйте, сейчас работаете? хотел бы арендовать шуруповерт', 'Добрый день! Нужен в прокат шуруповерт для сборки стола, подскажите стоимость и до скольки можно подойти забрать?'
 
Теперь сгенерируй {n_examples} обращений для темы: "{theme}"
"""

        response = self.call_deepseek(prompt, temperature=0.8)
        texts = response.split(';')

        return texts
```

```python
count_of_examples_needed = 10
df = pd.read_csv('../dataset/texts_from_avito.csv', sep=';')
df = df[df['тема'].notna()]
df = df[['тема', 'текст']]
theme_counts = df['тема'].value_counts()
rare_themes = theme_counts[theme_counts < count_of_examples_needed]
rare_themes
```

```python
deepseek = DeepSeekIntegration()

# тестовый запуск
theme = 'Отбойный молоток'
count_now = 9
n_examples = count_of_examples_needed #- count_now   
new_examples = deepseek.generate_texts_for_theme(theme=theme, n_examples=n_examples)
for new_example in new_examples:
    df.loc[len(df)] = [theme, new_example]
```

```python
df_small = df[df['тема']=='Отбойный молоток']
examples = df_small['текст'].tolist()
examples
```

```python
theme_counts = df['тема'].value_counts()
rare_themes = theme_counts[theme_counts < count_of_examples_needed]
rare_themes
```

```python
# полный запуск
import time

for theme, count_now in rare_themes.items():
    try:
        print(f'Получаю новые тексты для темы {theme}...')
        start = time.time()
        n_examples = count_of_examples_needed #- count_now   
        new_examples = deepseek.generate_texts_for_theme(theme=theme, n_examples=n_examples)
        time_spent = time.time() - start
        print(f'Заняло времени {time_spent}.')
        for new_example in new_examples:
            df.loc[len(df)] = [theme, new_example]
    except Exception as e:
        print(e)

df.to_csv('../dataset/extended_by_deepseek_3.csv')
```
