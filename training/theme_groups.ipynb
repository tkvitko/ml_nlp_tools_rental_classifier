```python
import pandas as pd
```

```python
df = pd.read_csv('../dataset/theme_groups2.csv', sep=';')
df = df[['тема', 'группа']]
df
```

```python
group_counts = df['группа'].value_counts()
group_counts
```

```python
# объединим группы с 1 темой в группу "прочее"
rare_groups = group_counts[group_counts == 1].index
df['группа'] = df['группа'].replace(rare_groups, 'прочее')
group_counts = df['группа'].value_counts()
group_counts
```
